# Scheduler CLI Examples

This notebook demonstrates using the `alma schedule` CLI commands.

```python
import subprocess

def run_cli(*args):
    """Helper to run CLI commands."""
    result = subprocess.run(
        ["uv", "run", "alma", *args],
        capture_output=True,
        text=True
    )
    print(result.stdout)
    if result.stderr:
        print("STDERR:", result.stderr)
    return result
```

## Basic Schedule Generation

Generate a lineup for today:

```python
run_cli("schedule", "generate", "--help")
```

## Natural Language Requests

Use the `--request` flag to specify what you want:

```python
# Generate with specific request
run_cli("schedule", "generate", "--request", "one blueie")
```

```python
# Multiple requests
run_cli("schedule", "generate", "--request", "one blueie and two throw throw")
```

## Date Specification

You can specify dates in multiple ways:

```python
# Using keywords
run_cli("schedule", "generate", "tomorrow")
```

```python
# Date from request text
run_cli("schedule", "generate", "--request", "tomorrow one blueie")
```

## View Generated Schedule

```python
# Show today's schedule
run_cli("schedule", "today")
```

```python
# JSON output for automation
run_cli("schedule", "today", "--json")
```
